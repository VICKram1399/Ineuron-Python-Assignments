```python
def myreduce(func, itr, init=None ):
    
    if init:
        val, iter = init, itr
    else:
        val, iter = itr[0], itr[1:]
    

    for i in iter:
        
        val = func(val, i)
        

    return val
## Testing the function created:

def ad(x,y):
    return x+y

myreduce(ad,[1,3,4,5,8,9])
```

```python
def myfilter(func,itr,init=None):
    if init:
        val1,val2,iter= init, init,itr
    else:
        val1,val2,iter= itr[0],itr[0],itr[1:]
    for i in iter:
        val1,val2= func(iter,i),func(iter,i)
    return val1,val2
```
