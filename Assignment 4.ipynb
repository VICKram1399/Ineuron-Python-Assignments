```python
## 1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
# formula.
##  area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
# Function to take the length of the sides of triangle from user should be defined in the parent
# class and function to calculate the area should be defined in subclass.


class sides:
    def __init__(self,a,b,c):
        self.a=float(a)
        self.b=float(b)
        self.c=float(c)
a=int(input("a="))
b=int(input("b="))
c=int(input("c="))

class area(sides):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)
        
    def the_area(self):
        s=(a+b+c) / 2
        return  (s*(s-a)*(s-b)*(s-c)) ** 0.5
    
    
t= area(a,b,c)
print("the area of the triange is= {}".format(t.the_area()))
            
```

```python
# 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns
##  the list of words that are longer than n.


def filter_long_words(ListOfWords,n):
    return(word for word in ListOfWords if len(word) >= n)
def myfun():
    ListOfWords=input("enter your words,with spaces=").split()
    n= int(input("enter min length="))
    print("List of words that are longer than={} are={}".format(n,','.join(filter_long_words(ListOfWords,n))))
    
myfun()
```

```python
# 2.2 : Write a Python function which takes a character (i.e. a string of length 1) and returns True if
##it is a vowel, False otherwise.

def vowelfunc(input_ch):
    
    for x in input_ch:
        if (x=="a" or x=="e" or x=="i" or x=="o" or x=="u"):
            return True
        else:
            return False
        
input_ch=input("enter your character=")

if vowelfunc(input_ch)==True:
    print("your character IS a Vowel")
else:
    print("your character is NOT a Vowel")
```
