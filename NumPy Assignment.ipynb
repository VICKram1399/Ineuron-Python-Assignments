```python
# Write a function to find moving average in an array over a window:
# Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.
```

```python
def movingavg(a,n=3):
    asum= np.cumsum(a) #Using the "Cumulative Sum" function of the NumPy package
    asum[n:]= asum[n:]- asum[:-n] #breaking the cumulative sum from the previos element
    return asum[n-1:]/n
```

```python
a= np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])
x= np.array(a)
movingavg(x,3).round(2) #rounding off the output decimal point to 2 
```
